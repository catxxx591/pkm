```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
pkm = pd.read_csv('pkm.csv')
list(pkm['Name'])
```

```python
#把第一欄拿掉
pkm.drop(pkm.columns[0],inplace=True,axis = 1)
#或者
#pkm.drop(pkm['index'],inplace=True)
pkm.head()
```

```python
pkm.groupby('Type 1').first()
#pkm.groupby(pkm['HP']>60).first()
```

```python
pkm.dtypes
```

```python
type1 = pkm.groupby('Type 1').mean().round()
type1
```

```python

type1 = type1.T
```

```python
type1['mean']=0
type1
```

```python
n=0
while n < len(type1['mean']):
    type1['mean'][n]= type1.loc[type1.index[n]].mean().round()
    n += 1
type1
```

```python
type1['mean'].plot()
#pkm.iloc[12,3:9].plot()

namee = 'Pikachu'
def who(name):
    type1['mean'].plot()
    pkm.loc[pkm['Name']== name,'HP':'Speed'].T[pkm.loc[pkm['Name']==name,'HP':'Speed'].T.columns[0]].plot()
    d = pkm.loc[pkm['Name']== name,'Type 1']
    d = d.values[0]
    type1[d].plot()
    print( pd.DataFrame(pkm.loc[pkm['Name']== name,'Name':'Speed']) )

#輸入名字 = Pikachu
#黃 : Pikachu的能力值
#綠 : Pikachu屬性(電)的平均能力值
#藍 : 全部精靈的平均能力值
```
